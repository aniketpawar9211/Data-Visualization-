```python
import seaborn as sns
```

```python
df=sns.load_dataset('tips')
```

```python
df.head()
```

# Correlation with heatmap (Applied only on Numerical data)

```python
df.corr()
```

```python
sns.heatmap(df.corr())
```

# Join Plot

```python
sns.jointplot(x='tip',y='total_bill',data=df)
```

```python
sns.jointplot(x='tip',y='total_bill',data=df,kind='hex')
```

```python
sns.jointplot(x='tip',y='total_bill',data=df,kind='reg')
```

# Pair Plot

```python
sns.pairplot(df)
```

```python
sns.pairplot(df,hue='smoker')
```

# Dist Plot

```python
sns.distplot(df['tip'])
```

```python
sns.distplot(df['tip'],kde=False)
```

# Analysis on Categorical Data

## Count Plot

```python
sns.countplot('sex',data=df)
```

```python
sns.countplot('smoker',data=df)
```

```python
sns.countplot('day',data=df)
```

```python
sns.countplot(y='sex',data=df)
```

## Bar plot

```python
sns.barplot(x='total_bill',y='sex',data=df)
```

```python
sns.barplot(x='sex',y='tip',data=df)
```

## Box Plot

```python
sns.boxenplot(x='sex',y='total_bill',data=df)
```

```python
sns.boxplot(x='sex',y='total_bill',data=df)
```

```python
sns.boxplot(x='total_bill',y='day',hue='smoker',data=df)
```

## Violine Plot

```python
sns.violinplot(x='total_bill',y='day',data=df)
```
